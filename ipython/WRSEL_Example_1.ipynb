# Introducing the package WRSEL

In this tutorial we will examine the use of Weighted Ranks Square Error Loss (WRSEL) as described in Wright, Stern and Cressie (2003) as a means of obtaining optimal point estimates under rank-based loss functions. The package can be found on the author's github page https://github.com/andrewzm/WRSEL. Below we will consider 3 independent random variables, and see what happens to the optimal estimate for different weight vectors. This tutorial is accompanied by an Rshine application, where the impact of varying weights can be assessed interactively.

### Underlying theory

WRSEL is used to compute optimal point estimates $\hat \lambda$ of hidden variables $\lambda$ using a loss function which penalises squared errors according to the marginal variable's *rank*. Using the notation in Wright, Stern and Cressie (2003), consider a generic model with distribution $f( y | \lambda)$ where $y$ denotes the observed data. The WRSEL is given by

$W(\lambda, \hat\lambda;c) = \sum_{k} c_{r(k)}(\lambda_k - \hat\lambda_k)^2$

where $r(k)$ identifies the rank of $\lambda_k$ and $c_{r(k)}$ is the associated weight. Note that the weight assigned to the squared error between the value for the $k$th parameter $\lambda_k$ and the estimate $\hat\lambda_k$ depends on the rank of $\lambda_k$. This is what distinguishes WRSEL from standard squared error loss methods.

It can be shown that, under this loss function, the optimal estimate of $\hat\lambda_k$is given by

$\hat\lambda_k = \frac{\sum_j E(\lambda_k | \lambda_k = \lambda_{(j)},y)c_jP(\lambda_k = \lambda_{(j)} | y)}{\sum_j c_jP(\lambda_k = \lambda_{(j)} | y)}$

which is, in general, intractable. The package `WRSEL` computes this is efficiently using a representative sample from the posterior $p(\lambda | y)$ in two stages. In the first stage, the quantites $E(\lambda_k | \lambda_k = \lambda_{(j)},y)$ and $P(\lambda_k = \lambda_{(j)} | y)$ are computed. This is computationally intensive, but can be easily parallelised since the computation of, say, $P(\lambda_k = \lambda_{(1)} | y)$ is independent from that of $P(\lambda_k = \lambda_{(2)} | y)$ (same for the expectation). The user is then provided with a function to which any desired weight vector can be applied. The optimal estimates for any weight vector are then effectively provided with no additional computational cost.


### Toy example

In this toy study we consider three independent Gaussian variables with standard deviation 1 and with expectation [-1,0,1].  We generate N = 10000 samples from these distributions and take these to be representative samples from our $f(\lambda | y)$. Note that in a real application, these samples would be from a high-dimensional posterior distribution which, in general, is non-Gaussian. 

```python
library(Matrix)
library(ggplot2)
library(WRSEL)
n <- 3                           # number of variables
N <- 10000                       # number of MCMC samples
X <- matrix(rnorm(n*N),n,N)      # generate samples
X <- X  + c(-1,0,1)              # set the first random variable to be Gaussian with mean 1 (and sd = 1)
```

The function `WRSEL`, which is the primary function in the package `WRSEL`, takes these samples as an input argument and returns a function with which WRSEL can be applied according to a user-specified weight vector. This preliminary step is the computationally intensive part -- once the function is computed and returned, the optimal estimates for different weight functions are obtained for free. The function `WRSEL` supports a parallel architecture since, as explained above, the computation of the conditional expectations and conditional probabilities are embarassingly parallel in the order statistic. Parallelisation guarantees a theoretically lineary speed-up, however in practice the actual speed-up obtained is dependent on the architecture. For this simple problem we will keep `num_nodes = NULL` to indicate that we do not wish to parallelise this operation.

```python
WRSEL_fun <- WRSEL(X = X,num_nodes = NULL)
```

`WRSEL_fun` is a function which can compute the optimal estimate for a given weight vector. This vector needs to be equal in length to the number of variables we are using, in this case 3. In this simple case study we assume that if we err on the low side, the loss is large. In fact we will take the extreme case of assuming that the loss vector is zero everywhere except for the largest order statistic. In this case we set:

```python
weights <- c(0,0,1)
```

which sets the optimal estimate equal to the expectation of the variable conditional on it being the maximum from all three. To evaluate the optimal estimate we simply run

```python
lambda_hat <- WRSEL_fun(weights)
print(lambda_hat)
```

Here `lambda_hat` is our estimate under the considered weight function. It is interesting to see how differs from the mean which, in this case, is [-1,0,1] and which would be the optimal estimate under a weight vector of the form `c[1,1,1]` where `c` is a constant. Plotting the resulting sample density and point estimates we see that these estimates are considerably different from the mean, and reflect the choice of loss function, i.e. that erring on the lower side would be more costly than erring on the high side. 

```python
df <- reshape2::melt(cbind(data.frame(t(X)),n=1:ncol(X)),id.vars = "n")
df2 <- reshape2::melt(cbind(data.frame(t(lambda_hat)),n=1:3),id.vars="n")
ggplot(df) + geom_density(aes(x=value,fill=variable),alpha=0.3) + 
  geom_vline(data=data.frame(df2),aes(xintercept=value,colour = variable))
```

Interestingly, although the variables are independent, the point estimates do not differ from the mean to the same extent. Whilst all the estimates are pushed up, those which have a low expectation are pushed higher than that with a higher expectation. Recall that this weight vector returns the expectation of $\lambda$ conditional on the marginals being the largest of the three. Therefore, the presence of samples from x3 being large, will inevitable cause this optimal estimate of x1 to also be quite large. This is why this setting is labelled a 'worst case scenario' in Wright, Stern and Cressie (2002).
